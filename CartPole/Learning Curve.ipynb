```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] 
```

```python
DQN = np.load('curve\DQN.npy')[:3000]
DAQN = np.load('curve\DAQN.npy')[:3000]
DATQN = np.load('curve\DATQN.npy')[:3000]
AC = np.load('curve\AC.npy')[:3000]
```

```python
plt.plot(DQN)
plt.plot(DAQN)
plt.plot(DATQN)
plt.xlabel("episode")
plt.ylabel("average mean of last 100 episodes")
plt.xlim(-100,3200)
#plt.title("Comparison of 3 Different Method")
plt.legend(["深度Q网络",'先后训练',"同时训练"],loc='lower right',ncol=1)
plt.savefig('CartPole.jpg',dpi = 200)
```
